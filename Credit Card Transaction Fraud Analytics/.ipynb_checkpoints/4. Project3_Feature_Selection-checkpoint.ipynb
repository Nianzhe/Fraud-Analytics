```python
import pandas as pd
import numpy as np
import random
```

```python
v = pd.read_csv('final_variable.csv')
```

```python
v[['Fraud','cardnum-merch_state_merch_zip_count30__mean']].info()
```

```python
v.columns
```

```python
v = v[(v.Recnum <=84299) & (v.Recnum >= 3338)]
```

```python
v = v.set_index('Recnum')
```

```python
from scipy import stats
```

```python
KSFDR = pd.DataFrame(columns = ['ks'])
```

```python
goods = v[v['Fraud'] == 0]
bads = v[v['Fraud'] == 1]
```

```python
i = 0
for column in list(v.columns):
    KSFDR.loc[i,'ks'] = stats.ks_2samp(goods[column],bads[column])[0]
    i = i+1
```

```python
KSFDR.index = v.columns
```

```python
KSFDR = KSFDR.reset_index()
```

```python
KSFDR.sort_values('ks',ascending = False)
```

```python
numbads = len(bads)
```

```python
topRows = int(round(len(v)*0.03))
j = 0
for column in v.columns:
    if column != 'Fraud':
        temp1 = v[[column,'Fraud']].sort_values(column,ascending=False).head(topRows)
        temp2 = v[[column,'Fraud']].sort_values(column,ascending=True).head(topRows)
        needed1 = temp1.loc[:,'Fraud']
        needed2 = temp2.loc[:,'Fraud']
        FDR1 = sum(needed1)/numbads
        FDR2 = sum(needed2)/numbads
        FDRate = np.maximum(FDR1,FDR2)
        KSFDR.loc[j, 'FDR'] = FDRate
        j = j +1
    else:
        KSFDR.loc[j, 'FDR'] = 1
        j = j+1
```

```python
KSFDR['KS_rank'] = KSFDR['ks'].rank(axis = 0)
KSFDR['FDR_rank'] = KSFDR['FDR'].rank(axis = 0)
KSFDR['Average_rank'] = (KSFDR['KS_rank'] + KSFDR['FDR_rank'])/2
```

```python
KSFDR.sort_values(['Average_rank']).tail(81)
```

```python
KSFDR.sort_values(['Average_rank']).tail(81).to_excel('KS_FDR.xlsx')
```

```python
df=v.copy()
```

```python
df=df[KSFDR.sort_values(['Average_rank']).tail(81)['index']]
```

```python
df=df.drop(['Fraud'],axis=1)
```

```python
len(df.columns)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(max_iter = 1000)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
data=scaler.fit_transform(df.to_numpy())
data=pd.DataFrame(data,columns=df.columns)
```

```python
for col in list(data.columns):
    a = np.array(data[col].values.tolist())
    data[col] = np.where(a > 9, 10, a).tolist()
```

### 1.1 Wraper - Logistic Regression

```python
from sklearn.feature_selection import RFECV
```

```python
rfecv=RFECV(estimator=model,step=1,cv=3,verbose=1,n_jobs=1,scoring='roc_auc')
rfecv.fit(data,v['Fraud'])
```

```python
var_selected = pd.DataFrame(sorted(zip(map(lambda x: round(x),rfecv.ranking_),data.columns)),columns = ['ranking', 'variable'])
var_selected.head(25)
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1,len(rfecv.grid_scores_)+1),rfecv.grid_scores_)
plt.xlabel('Number of features selected')
plt.ylabel('Cross validation score (nb of correct classifications)')
plt.savefig('score1.png')
```

```python
var_selected.to_csv('selected_1.csv')
```

### 1.2 Wrapper - second step

```python
select_column = var_selected[var_selected['ranking'] <= 50].variable.to_list()
```

```python
data2 = data[select_column]
```

```python
rfecv2=RFECV(estimator=model,step=1,cv=3,verbose=1,n_jobs=1,scoring='roc_auc')
rfecv2.fit(data2,v['Fraud'])
```

```python
var_selected2 = pd.DataFrame(sorted(zip(map(lambda x: round(x),rfecv2.ranking_),data2.columns)),columns = ['ranking', 'variable'])
var_selected2.head(25)
```

```python
var_selected2.to_csv('selected_2.csv')
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1,len(rfecv2.grid_scores_)+1),rfecv2.grid_scores_)
plt.xlabel('Number of features selected')
plt.ylabel('Cross validation score (nb of correct classifications)')
plt.savefig('score2.png')
```
