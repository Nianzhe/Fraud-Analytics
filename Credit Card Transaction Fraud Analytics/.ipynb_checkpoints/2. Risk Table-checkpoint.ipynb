```python
import pandas as pd
import numpy as np
import datetime
```

```python
df=pd.read_excel('no_missing.xlsx', converters={'Merchnum': lambda x: str(x), 'Merch_zip': lambda x : str(x)})
```

```python
df.head()
```

```python
df['dayofweek']=df['Date'].dt.dayofweek
day_week=df[['Fraud','dayofweek']][df['Recnum']<=84299].groupby(by='dayofweek').mean().reset_index()
day_week.columns=['dayofweek','day_week_risk']
```

```python
day_week
```

```python
df=pd.merge(df,day_week,left_on='dayofweek',right_on='dayofweek',how='left')
```

```python
grouprisk=df[['Recnum','day_week_risk']].set_index('Recnum')
```

```python
grouprisk
```

```python
grouprisk.to_excel('Risk Table.xlsx')
```
