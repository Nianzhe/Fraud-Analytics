```python
train_data = df.iloc[:83970]
variables = list(train_data.columns)
```

```python
variables.remove("Recnum")
```

```python
# Calculating KS
from scipy import stats 
KS = []
for v in variables:
    if v!="Fraud":
        tempt_df = train_data[[v,"Fraud"]]
        col_bad = tempt_df[tempt_df["Fraud"]==1][v]
        col_good = tempt_df[tempt_df["Fraud"]==0][v]
        KS_score = stats.ks_2samp(col_bad,col_good)[0]
        KS.append(KS_score)
    else:
        tempt_df = train_data[["Fraud"]]
        col_bad = tempt_df[tempt_df["Fraud"]==1]['Fraud']
        col_good = tempt_df[tempt_df["Fraud"]==0]['Fraud']
        KS_score = stats.ks_2samp(col_bad,col_good)[0]
        KS.append(KS_score)
```

```python
# Calculating FDR
FDR = []
FDR_3_percent= int(train_data.shape[0]*0.03)
all_frauds_count = train_data[train_data["Fraud"]==1].shape[0]
for col in variables:
    if col!="Fraud":
        tempt_df1 = train_data[[col,'Fraud']].sort_values(ascending=False,by=[col]).head(FDR_3_percent)
        tempt_df2 = train_data[[col,'Fraud']].sort_values(ascending=True,by=[col]).head(FDR_3_percent)
        tempt_bad1 = tempt_df1[tempt_df1["Fraud"]==1].shape[0]
        tempt_bad2 = tempt_df2[tempt_df2["Fraud"]==1].shape[0]
        FDR_rate = np.maximum(tempt_bad1/all_frauds_count,tempt_bad2/all_frauds_count)
        FDR.append(FDR_rate)
    else:
        tempt_df = train_data[[col]].sort_values(ascending=False,by=[col]).head(FDR_3_percent).astype(int)
        FDR_rate = sum(tempt_df.loc[:,"Fraud"])/all_frauds_count
        FDR.append(FDR_rate)
```

```python
data = {"KS":KS,"FDR":FDR,"index":variables}
KSFDR = pd.DataFrame(data)
KSFDR.
KSFDR.set_index("index")
```
