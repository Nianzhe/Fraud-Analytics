```python
independent = df[selected]
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
data=scaler.fit_transform(independent.to_numpy())
data=pd.DataFrame(data,columns=independent.columns)
```

```python
for col in list(data.columns):
    a = np.array(data[col].values.tolist())
    data[col] = np.where(a > 5, 5, a).tolist()
```

```python
data['Recnum'] = df['Recnum']
data['Fraud'] = df['Fraud']
```
